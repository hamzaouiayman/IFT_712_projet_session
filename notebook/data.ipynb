```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import itertools as iter
```

```python
file_train = json.load(open("../Data/Raw/train.json"))
file_test = json.load(open("../Data/Raw/train.json"))
```

```python
df_train = json_normalize(file_train)
df_test = json_normalize(file_test)
```

```python
df_train = df_train[['cuisine','ingredients']]
```

```python
ingredients_train = df_train['ingredients'].tolist()
ingredients_test = df_test['ingredients'].tolist()
```

```python
ingredients_train = list(iter.chain.from_iterable(ingredients_train)) 
ingredients_test = list(iter.chain.from_iterable(ingredients_test))
```

```python
ingredients_train = list(set(ingredients_train))
ingredients_test = list(set(ingredients_test))
```

```python
print(len(ingredients_train))
print(len(ingredients_test))
```

```python
cuisine = df_train.cuisine.unique()
```

```python
DF_train = df_train['ingredients'].apply(lambda x: pd.Series(x)).stack().reset_index(level=1, drop=True).to_frame('ingredients').join(df_train[['cuisine']], how='right' )
```

```python
pd.crosstab(DF_train['cuisine'],DF_train['ingredients'])
```

```python
DF_test = df_test['ingredients'].apply(lambda x: pd.Series(x)).stack().reset_index(level=1, drop=True).to_frame('ingredients').join(df_test[['id']], how='right' )
```

```python
pd.crosstab(DF_test['id'],DF_test['ingredients'])
```
